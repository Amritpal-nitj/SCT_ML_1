**PREDICT THE PRICE OF HOUSE IN BANGLORE**

USING LINEAR REGRESSION


```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB"
df = pd.read_csv(path)
```

```python
df.head()
```

SPLIT DATA


```python

X=df.drop("price",axis=1)  #gives data except price
#X is capital,(independent) because it represents matrix while y is a vector,(dependent)
y=df["price"]
X.head()
```

```python
y.head()
```

```python
print("shape of X:",X.shape)
print("shape of y:",y.shape)
```

**train_test_split() is a scikit-Learn class help to Split data in
to 4 dataset (like: X_train, X_test, y_train, y_test.) **

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51) #20%testing ke liye baaki 80% for training of model
#We have used random to fix our training data
print("sahpe of X_train:",X_train.shape)
print("sahpe of y_train:",y_train.shape)
print("sahpe of X_test:",X_test.shape)
print("sahpe of y_test:",y_test.shape)
```

#FEATURE SCALING
## we need to scale independent to train the model perfectly it normalizes the values across all data between 0 and 1

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)

#No need to transform dependent
```

```python
X_test
```

LINEAR RIGRESSION : ML MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_   #b1 b2 ....b107   #jo jo learn kiya model ne
```

```python
lr.intercept_  #b0
```

PREDICt THE VALUE(PRIZE) OF HOME AND TEST

```python
X_test[0,:]
```

```python
lr.predict([X_test[0,:]])    #tells the prze of home 76 lakh
```

```python
X_test
```

```python
lr.predict(X_test)
```

```python
y_test  #to verify our models test (some values are matching nearly, some are not)
```

TO CHEAK ACCURACY (comparing x test and y test)

```python
lr.score(X_test,y_test)   #79% ACCURACY
```

# **MODEL EVALUATION**

ROOT MEAN SQUARE ERROR

```python
y_predict=lr.predict(X_test)
y_predict
```

```python
y_test
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_predict)
rmse=np.sqrt(mse)
print(" mean square error:",mse)
print("root mean square error:",rmse)
```
